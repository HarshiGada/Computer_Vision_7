```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

from collections import Counter
```

```python
# k=3 to largest
```

```python
img1 = cv2.imread('img1.jpg')
```

```python
image_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
plt.imshow(image_rgb)
plt.title('Original Img')
```

```python

pixels = image_rgb.reshape(-1, 3)
# Dynamically adjust subplot grid
num_clusters = range(1, 10)
rows = (len(num_clusters) + 1) // 2  # Ensure enough rows
cols = 2  # Keep 2 columns for display

# Create a figure for displaying multiple segmented images
plt.figure(figsize=(15, 15))

# Iterate over different values of k
for i, k in enumerate(num_clusters):
    # Apply K-means clustering
    kmeans = KMeans(n_clusters=k, random_state=42, max_iter=500)
    kmeans.fit(pixels)

    # Get the cluster centers (the RGB values of the centroids)
    cluster_centers = kmeans.cluster_centers_

    # Get the labels (which cluster each pixel belongs to)
    labels = kmeans.labels_

    # Replace each pixel value with the color of its corresponding cluster center
    segmented_image = cluster_centers[labels].reshape(image_rgb.shape)

    # Convert back to uint8 for proper display
    segmented_image = np.uint8(segmented_image)

    # Plot the segmented image
    plt.subplot(rows, cols, i + 1)  # Adjust dynamically
    plt.title(f"Segmented Image with k={k}")
    plt.imshow(segmented_image)
    plt.axis('off')



```

Conclusion: K means algorithm is used to segment the image (flower field image).

For k=1,it shows average value of entire image including rgb values.
the image has constant value which is equal to mean of all pixels of all channels.

As K value increase the number of segment increases till k=8. This value of k is equal to no of possible pixel value/colors in the given image.
